```python
import numpy as np
import pandas as pd
```

```python
bar_locations = pd.read_csv("bar_locations.csv")
party_in_nyc = pd.read_csv("party_in_nyc.csv")
```

```python
np_bar = np.array(bar_locations)
np_bar
```

```python
np_party = np.array(party_in_nyc)
np_party
```

```python
combined_data = []
latitudes = []
longitudes = []
for b in np_bar:
    for p in np_party:
        if p[-2] == b[-3] and p[-1] == b[-2]:
            if not (p[-2] in latitudes and p[-1] in longitudes):  
                s = np.hstack([b, p])
                combined_data.append(s)
                latitudes.append(p[-2])
                longitudes.append(p[-1])
        
```

```python
isoToUtc = lambda date, _: (datetime.strptime(date, '%Y-%m-%d %H:%M:%S') - datetime(1970,1,1)).total_seconds()
```

```python
np_isoToUtc = np.vectorize(isoToUtc);
```
